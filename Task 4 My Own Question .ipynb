# HOME

```python
df_hw = df[df["home_team_result"]=="Win"]
df_hw
```

```python
#FRIENDLY HOME WINS
df_friendly = df_hw[df_hw['tournament']=='Friendly']
panchod = df_friendly.groupby('home_team').agg({'home_team_result': 'count'})
panchod.sort_values(by = 'home_team_result', ascending = False).head(10)
```

```python
#TOTAL HOME WINS
anhad = df_hw.groupby('home_team').agg({'home_team_result': "count"})
anhad.sort_values(by ='home_team_result', ascending = False).head(10)
```

```python
#TOTAL WINS by WC Quali's (home-team) 
df_quali = df_hw[df_hw['tournament']== "FIFA World Cup qualification"]
quali = df_quali.groupby('home_team').agg({'home_team_result': 'count'})
quali.sort_values(by = 'home_team_result', ascending = False).head(10)
```

```python
#World Cup elim wins (home)
df_wc = df_hw[df_hw['tournament']== "FIFA World Cup"]
wc = df_wc.groupby('home_team').agg({'home_team_result': 'count'})
wc.sort_values(by = 'home_team_result', ascending = False).head(10)
```

```python
group_names = ["Win", "Lose", "Draw"]
```

```python
#adding away team result column 
def condition(x):
    if x == "Lose":
        return group_names[0]
    elif x == "Win":
        return group_names[1]
    else:
        return group_names[2]
df["away team result"] = df["home_team_result"].map(condition)
df
```

# AWAY

```python
df_aw = df[df["away team result"]=="Win"]
```

```python
#FRIENDLY
df_friend = df_aw[df_aw['tournament']=='Friendly']
hello = df_friend.groupby('away_team').agg({'away team result': 'count'})
hello.sort_values(by = 'away team result', ascending = False).head(10)
```

```python
#TOTAL AWAY WINS
total_away = df_aw.groupby('away_team').agg({'away team result': "count"})
total_away.sort_values(by ='away team result', ascending = False).head(10)
```

```python
#TOTAL WINS by WC Quali's (away-team) 
df_quali_a = df_aw[df_aw['tournament']== "FIFA World Cup qualification"]
quali_a = df_quali_a.groupby('away_team').agg({'away team result': 'count'})
quali_a.sort_values(by = 'away team result', ascending = False).head(10)
```

```python
#World Cup elim wins (away)
df_wc_a = df_aw[df_aw['tournament']== "FIFA World Cup"]
wc_a = df_wc_a.groupby('away_team').agg({'away team result': 'count'})
wc_a.sort_values(by = 'away team result', ascending = False).head(10)
```
