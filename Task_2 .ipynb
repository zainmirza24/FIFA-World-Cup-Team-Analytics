```python
import pandas as pd
import numpy as np

df = pd.read_csv('/Users/Mirzafamily/Desktop/DS311-GroupProject/sport_data_description_tasks/fifa_worldcup.csv')
```

# Summary

```python
#NA Report
ls = list(df.columns)
for x in ls:
  if sum(df[x].isna()) > 0:
    print(x)
    print(sum(df[x].isna()))
    print()
```

```python
df.info()
```

```python
df.groupby('neutral_location')['country'].count() #T/F games played in neutral location 
```

```python
#top home team mean goals scored
goals_h = df.groupby('home_team')['home_team_score'].mean()
df_goalsH = goals_h.to_frame()
df_goalsH.sort_values(by = 'home_team_score', ascending = False).head(20)
```

```python
homeO = df.groupby('home_team')['home_team_mean_offense_score'].mean()
df_homeO = homeO.to_frame()
df_homeO.sort_values(by = 'home_team_mean_offense_score', ascending = False).head(10)
#TOP HOME TEAM OFFENSE 
```

```python
goals_a = df.groupby('away_team')['away_team_score'].mean()
df_goalsA = goals_a.to_frame()
df_goalsA.sort_values(by = 'away_team_score', ascending = False).head(10)
```

```python
awayO = df.groupby('away_team')['away_team_mean_offense_score'].mean()
df_awayO = awayO.to_frame()
df_awayO.sort_values(by = 'away_team_mean_offense_score', ascending = False).head(10)
#TOP AWAY TEAM OFFENSE 
```

# Away teams with their corresponding average away defense score

```python
awayd = df.groupby('away_team')['away_team_mean_defense_score'].mean() 
df_away = awayd.to_frame() #Series to Dataframe
```
